```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#import requests
import time
import ast

# Incorporated citipy to determine city based on latitude and longitude
#from citipy import citipy
```

```python
cit = pd.read_csv("CityWeather.csv")
cit
```

```python
#html for web of data
html = cit.to_html('data.htm')
print(html)
```

```python
ast.literal_eval(cit['coord'][0])
coord_dict = {}
i = 0
city_ls = []
lat_ls = []
lon_ls = []
for city in cit['coord']:
    coord_line = ast.literal_eval(cit['coord'][i])
    city = cit['Unnamed: 0'][i]
    coord_dict[city] = coord_line
    city_ls.append(city)
    lat_ls.append(coord_line['lat'])
    lon_ls.append(coord_line['lon'])
    i += 1
coord_dict
```

```python
lat_df = pd.DataFrame(coord_dict)
lat_df = pd.DataFrame.transpose(lat_df)
lat_df
```

```python
plt.scatter(lat_df['lon'], lat_df['lat'])
plt.title("Longitudes / Latitudes")
plt.ylabel('Latitudes')
plt.xlabel("Longitudes")
plt.grid(True)
# Save the Figure
plt.savefig('Images/Map.png')
# Show the Figure
plt.show()
```
